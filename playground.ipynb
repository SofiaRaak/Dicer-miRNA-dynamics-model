```python
import model
import params
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
res1 = solve_ivp(model.toy_model, (0, int(params.minutes)), params.init_values)
```

```python
dicer, premirna1, premirna2, premirna3, premirna4, premirna5, premirna1_dicer, \
premirna2_dicer, premirna3_dicer, premirna4_dicer, premirna5_dicer, mirna1, mirna2, \
mirna3, mirna4, mirna5 = res1.y
```

```python
fig, ax = plt.subplots()

ax.plot(res1.t, mirna1)
ax.plot(res1.t, mirna2, linestyle = '-.')
ax.plot(res1.t, mirna3, linestyle = '--')
ax.plot(res1.t, mirna4, linestyle = ':')
ax.plot(res1.t, mirna5, alpha = 0.5)
#ax.plot(res1.t, dicer)

ax.legend(['miR-1', 'miR-2', 'miR-3', 'miR-4', 'miR-5'])

ax.set_xlabel('Time (minutes)')
ax.set_ylabel('MiRNA concentration (nM)')

plt.show()
```

```python
fig = plt.figure()
plt.rc('font', size=12)
plt.rcParams['svg.fonttype'] = 'none'

ax1 = plt.subplot(2,3,1)
ax1.plot(res1.t, mirna1, color = 'C0')
ax1.set_ylabel('miR-1 (nM)')
ax1.set_xlabel('Time (min)')

ax2 = plt.subplot(2,3,2, sharex=ax1, sharey=ax1)
ax2.plot(res1.t, mirna2, color = 'C1')
ax2.set_ylabel('miR-2 (nM)')
ax2.set_xlabel('Time (min)')

ax3 = plt.subplot(2,3,3, sharex=ax1, sharey=ax1)
ax3.plot(res1.t, mirna3, color = 'C2')
ax3.set_ylabel('miR-3 (nM)')
ax3.set_xlabel('Time (min)')

ax4 = plt.subplot(2,3,4, sharex=ax1, sharey=ax1)
ax4.plot(res1.t, mirna4, color = 'C3')
ax4.set_ylabel('miR-4 (nM)')
ax4.set_xlabel('Time (min)')

ax5 = plt.subplot(2,3,5, sharex=ax1, sharey=ax1)
ax5.plot(res1.t, mirna5, color = 'C4')
ax5.set_ylabel('miR-5 (nM)')
ax5.set_xlabel('Time (min)')

ax6 = plt.subplot(2,3,6)
ax6.plot(res1.t, dicer, color = 'C5')
ax6.set_ylabel('Dicer (nM)')
ax6.set_xlabel('Time (min)')

plt.tight_layout()

fig.savefig(r'./figures/toy_model.svg', dpi = 600, bbox_inches = 'tight')

plt.show()
```
